## Data Visualization With Matplotlib

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Read the data using pandas
cancer=pd.read_csv('data.csv')
cancer
```

It can be read from the table above that there are 569 patient data.

```python
#Check out the tail of the dataframe
cancer.tail(3)
```

```python
#Read radius data
k=cancer['radius_mean']
k
```

```python
#Read area data
l=cancer['area_mean']
l
```

```python
#Plot basic line plot
cancer.plot(x='area_mean', y='radius_mean', figsize=(10,5), color='crimson', linewidth='0.5')
plt.ylabel('')
plt.title('Relationship between area and radius')
plt.legend(loc='lower right')
plt.grid()
```

```python
#Plot scatterplot
x=cancer['perimeter_mean']
y=cancer['radius_mean']
plt.figure(figsize=(10,5))
plt.scatter(x,y, color='peachpuff', linewidth='0.5')
plt.grid()
```

```python
#Plot a graph containing values of concavitiy, concave points and smoothness
cancer.plot(x='diagnosis', y=['concavity_mean','smoothness_mean', 'compactness_mean'], figsize=(10,7))
plt.ylabel('')
plt.legend(loc='upper left')
plt.grid()
```

```python
#Plot subplots 1
plt.figure(figsize=(10,7))
plt.subplot(3,1,1)                   
plt.plot(cancer['perimeter_mean'],'b--')
plt.ylabel('Perimeter')
plt.grid()

plt.subplot(3,1,2)
plt.plot(cancer['radius_mean'], 'r.')
plt.ylabel('Radius')
plt.grid()

plt.subplot(3,1,3)
plt.plot(cancer['concavity_mean'], 'y*')
plt.ylabel('Concavity')
plt.grid()
```

```python
#Plot subplots 2
plt.figure(figsize=(10,7))
plt.subplot(1,3,1)
plt.plot(cancer['perimeter_mean'],'b*')
plt.title('Perimeter')
plt.grid()

plt.subplot(1,3,2)
plt.plot(cancer['radius_mean'], 'g--')
plt.title('Radius')
plt.grid()

plt.subplot(1,3,3)
plt.plot(cancer['concavity_mean'], 'r.')
plt.title('Concavity')
plt.grid()
```

```python
#Plot 3D plot
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure(figsize=(7,7))
ax=fig.add_subplot(111, projection='3d')

cancer=pd.read_csv('data.csv')
x=cancer['radius_mean'].tolist()
y=cancer['perimeter_mean'].tolist()
z=cancer['area_mean'].tolist()

ax.scatter(x,y,z, c='salmon', s=300)
ax.set_xlabel('X label')
ax.set_xlabel('Y label')
ax.set_xlabel('Z label')
```

```python
fig=plt.figure(figsize=(7,7))
ax=fig.add_subplot(111, projection='3d')

x=cancer['concavity_mean'].tolist()
y=cancer['perimeter_mean'].tolist()
z=cancer['area_mean'].tolist()

ax.scatter(x,y,z, c='crimson', s=300)
ax.set_xlabel('X label')
ax.set_xlabel('Y label')
ax.set_xlabel('Z label')
```

## Data Visualization With Seaborn 

```python
#Import seaborn library
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
cancer
```

```python
df_cancer=pd.DataFrame(np.c_[cancer['data'],cancer['target']],columns=np.append(cancer['feature_names'],['target']))
```

```python
df_cancer
```

```python
df_cancer.head(3)
```

```python
df_cancer.tail(5)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='mean area', y='mean smoothness', data=df_cancer)
```

The difference between mean area and mean smoothness is not seen in the above graph. We can get this visibility by adding `hue='target'` code.

```python
#Plot scatterplot
plt.figure(figsize=(10,5))
sns.scatterplot(x='mean area', y='mean smoothness', hue='target', data=df_cancer)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='mean area', y='mean radius', hue='target', data=df_cancer)
```

```python
#Plot countplot
plt.figure(figsize=(9,5))
sns.countplot(df_cancer['target'], label='Count')
```

In the histogram above, it can be seen how many samples belong to classes 0 and 1.

```python
#Plot pairplot
sns.pairplot(df_cancer, hue='target', vars=['mean radius', 'mean texture', 'mean area', 'mean perimeter','mean smoothness'] )
```

```python
#Plot heatmap
plt.figure(figsize=(20,20))
sns.heatmap(df_cancer.corr(), annot=True)
```

```python
#Plot distplot
sns.distplot(df_cancer['mean radius'], bins=25, color='greenyellow')
```

```python
#Plot two seperate distplot for target classes 0 and 1
class_0_df=df_cancer[df_cancer['target']==0]
class_1_df=df_cancer[df_cancer['target']==1]
```

```python
class_0_df
```

```python
class_1_df
```

```python
plt.figure(figsize=(10,7))
sns.distplot(class_0_df['mean radius'], bins=25, color='teal')
sns.distplot(class_1_df['mean radius'], bins=25, color='hotpink')
plt.grid()
```
